```python
# A linear regression example
# forward pass: x=1, y=2, w=1
# y = w * x, loss = (y_pred - y) ** 2

import torch

x, y, w = torch.tensor(1.0), torch.tensor(2.0), torch.tensor(1.0, requires_grad=True)
y_pred = x * w
loss = (y_pred - y) ** 2
print(loss)

loss.backward()
print(w.grad)
```
