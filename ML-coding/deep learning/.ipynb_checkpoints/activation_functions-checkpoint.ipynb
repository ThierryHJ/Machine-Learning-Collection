## softmax

<div>

</div>

<div>

</div>


```python
import numpy as np
import torch

x = np.array([2.0, 1.0, 0.1])

def softmax(array):
    prob_sum = np.sum(np.exp(array))
    return np.exp(x) / prob_sum

print(softmax(x))
```

```python
x = torch.tensor([2.0, 1.0, 0.1])
outputs = torch.softmax(x, axis=0)
print(outputs)
```

## Sigmoid

<div>

</div>

```python
y = np.array([2.0, 1.0, 0.1])

def sigmoid(y):
    return 1 / (1 + np.exp(-y))
    
print(sigmoid(y))
```

```python
y = torch.tensor([2.0, 1.0, 0.1])
print(torch.sigmoid(y))
```

## Tanh

<div>

</div>

```python
y = np.array([2.0, 1.0, 0.1])

def tanh(y):
    return 2 / (1 + np.exp(-2 * y)) - 1
    
print(tanh(y))
```

```python
y = torch.tensor([2.0, 1.0, 0.1])
print(torch.tanh(y))
```

## Relu

<div>

</div>


```python
y = np.array([2.0, 1.0, -0.1])

def relu(y):
    # return [i if i > 0 else 0 for i in y]
    return np.where(y > 0, y, 0)
print(relu(y))
```

```python
y = torch.tensor([2.0, 1.0, -0.1])
print(torch.relu(y))
```
