```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# define data, with sample of y = 2x
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=5, random_state=43)
# X = torch.from_numpy(X.astype(np.float32))
# y = torch.from_numpy(y.astype(np.float32))

# define weights and bias
n_samples, n_features = X.shape
weights = np.zeros(n_features)
bias = 0

# define lr, batch_size, etc
lr = 0.005
batch_size = 2
epoch = 10
num_iter = n_samples//batch_size * epoch
print(num_iter)
    

# for each batch/epoch
for i in range(num_iter):
    ## calculate y_pred
    y_pred = np.dot(X, weights) + bias

    ## calcualte loss: (y_pred - y) ** 2
    loss = np.mean(1/2 * (y_pred - y) ** 2)
    
    ## loss backward, update gradient d(loss)/dx = d(loss)/dy * dy/dw = np.dot(X.T * (y_pred - y))
    dw = 1 / n_samples * np.dot(X.T, (y_pred - y))
    db = 1 / n_samples * np.sum(y_pred - y)
    
    ## update weights
    weights -= lr * dw
    bias -= lr * db

    if (i + 1) % 100 == 0:
        print(i, loss)

# predict
y_pred = X * weights + bias
plt.plot(X, y, 'ro')
plt.plot(X, y_pred, 'b')
plt.show()
```

```python
### with pytorch
import torch.nn as nn

# X, y = datasets.make_regression(n_samples=100, n_features=1, noise=5, random_state=43)
X = torch.from_numpy(X.astype(np.float32))
y = torch.from_numpy(y.astype(np.float32)).reshape(y.shape[0], 1)
print(y.shape)


lr = 0.005
output_size = 1
model = nn.Linear(n_features, output_size)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

# for each batch/epoch
for i in range(num_iter):
    ## calculate y_pred
    y_pred = model(X)

    ## calcualte loss: (y_pred - y) ** 2
    loss = criterion(y_pred, y)
    
    ## loss backward, update gradient d(loss)/dx = d(loss)/dy * dy/dx = np.dot(X.T * (y_pred - y))
    loss.backward()
    
    ## update weights
    optimizer.step()

    optimizer.zero_grad()

    if (i + 1) % 100 == 0:
        print(i, loss)

# predict
y_pred = model(X).detach().numpy()
plt.plot(X, y, 'ro')
plt.plot(X, y_pred, 'b')
plt.show()
```
